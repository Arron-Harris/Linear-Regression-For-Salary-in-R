ggplot is a viualization library. Here we are plotting the age and monthly salary from the salary csv. we will set it equal to p.

```python
%%R
p <- ggplot(
  salary_data,
  aes(x=Age, y=Monthly.Salary)
) + geom_point(color='darkorange')
```

From the graphthese two variables have a positive correlation with eachother.

```python
%%R
p
```

Use a linear regression to plot the line on the graph.

```python
%%R
predicted_line <- data.frame(
  ages = c(20, 30, 40, 50, 60)
)
```

```python
%%R
predicted_line$predicted_salary <- (
  1000 + predicted_line$ages*500
)
```

visualizing the predicted line.

```python
%%R
ggplot(
  data = predicted_line,
  aes(x = ages, y= predicted_salary)
) + geom_line(color='red')
```

Line is off so we will have to adjust it 

```python
%%R
 p + geom_line(data = predicted_line,
              aes(ages, predicted_salary,
                  color='red')) 
```

```python
%%R
predicted_line$predicted_salary <- (
  -9000 + predicted_line$ages*500
)
```

The line pass through the middle of the graph. Confirming the positive correaltion and the relationship between the two.

```python
%%R
p + geom_line(data = predicted_line,
              aes(ages, predicted_salary,
                  color='red'))
```
